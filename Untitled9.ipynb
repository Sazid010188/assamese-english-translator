```python
# @title 🇮🇳 Assamese to English Translator 🇺🇸

# Install required packages
!pip install -q transformers gradio torch

import gradio as gr
from transformers import pipeline

print("🚀 Loading translation model... This may take a minute...")

# Load the translation model
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-mul-en")

def translate_assamese_to_english(assamese_text):
    """
    Translate Assamese text to English
    """
    if not assamese_text.strip():
        return "⚠️ Please enter some Assamese text"

    try:
        # Perform translation
        result = translator(assamese_text, max_length=128)[0]['translation_text']
        return result
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Create examples
examples = [
    "মোৰ নাম জন",
    "আজি বৰষুণ দিব",
    "তুমি কেমন আছা?",
    "আমি স্কুললৈ যাওঁ",
    "ভাৰত এখন মহান দেশ"
]

# Create the interface
print("📱 Creating user interface...")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # 🇮🇳 Assamese to English Translator 🇺🇸
        **Type any Assamese text below and get instant English translation!**
        """
    )

    with gr.Row():
        with gr.Column():
            input_text = gr.Textbox(
                label="📝 Assamese Text",
                placeholder="Type your Assamese text here...",
                lines=3
            )
            translate_btn = gr.Button("✨ Translate", variant="primary")

        with gr.Column():
            output_text = gr.Textbox(
                label="🔄 English Translation",
                lines=3,
                interactive=False
            )

    gr.Markdown("### 📚 Example Texts")
    gr.Examples(
        examples=examples,
        inputs=input_text,
        outputs=output_text,
        fn=translate_assamese_to_english,
        cache_examples=True
    )

    # Connect the button
    translate_btn.click(
        fn=translate_assamese_to_english,
        inputs=input_text,
        outputs=output_text
    )

    # Also work when pressing Enter
    input_text.submit(
        fn=translate_assamese_to_english,
        inputs=input_text,
        outputs=output_text
    )

print("✅ Ready! Launching translator...")
demo.launch(share=True, debug=True)
```
